```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
titanic = sns.load_dataset('titanic')

# Convert 'survived' to string (fix for legend bug)
titanic['survived'] = titanic['survived'].astype(str)

# Plot
sns.countplot(x='sex', hue='survived', data=titanic)
plt.title("Survival by Gender")
plt.legend(title='Survived')
plt.show()
```

```python
titanic = sns.load_dataset('titanic')
titanic['survived'] = titanic['survived'].astype(str)  # Fix for all hue=survived plots
```

```python
sns.countplot(x='class', hue='survived', data=titanic)
plt.title("Survival by Class")
plt.legend(title='Survived')
plt.show()
```

```python
sns.boxplot(x='survived', y='age', data=titanic)
plt.title("Age Distribution by Survival")
plt.show()
```

```python
# Create family size feature
titanic['family_size'] = titanic['sibsp'] + titanic['parch']

sns.histplot(data=titanic, x='family_size', hue='survived', bins=10, kde=True)
plt.title("Survival by Family Size")
plt.show()

```

```python
# Only include numerical columns
corr = titanic[['survived', 'age', 'fare', 'sibsp', 'parch']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
